```python
# Importando as bibliotecas necessÃ¡rias 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Estilo dos grÃ¡ficos
sns.set_style("whitegrid")
```

```python
# Carregando o dataset
df = pd.read_csv("C:\projeto_carrinho\.ipynb_checkpoints\carrinhos_abandonados.csv.csv")
```

```python
# VisualizaÃ§Ã£o inicial
display(df.head())
print(df.info())
```

```python
# Contagem total de registros
total = len(df)
```

```python
# Quantidade de compras e abandonos
compras = len(df[df['comprou'] == 'Sim'])
abandonos = len(df[df['comprou'] == 'NÃ£o'])
```

```python
# Taxa de abandono
taxa_abandono = abandonos / total * 100
print(f"ðŸ“‰ Taxa de abandono: {taxa_abandono:.2f}%")
```

```python
# Produto mais abandonado
produto_abandonado = df[df['comprou'] == 'NÃ£o']['produto'].value_counts().idxmax()
print(f"ðŸ“¦ Produto mais abandonado: {produto_abandonado}")
```

```python
# Categoria mais abandonada
categoria_abandonada = df[df['comprou'] == 'NÃ£o']['categoria'].value_counts().idxmax()
print(f"ðŸ›’ Categoria mais abandonada: {categoria_abandonada}")
```

```python
# HorÃ¡rio com mais abandonos
hora_abandono = df[df['comprou'] == 'NÃ£o']['horario'].value_counts().idxmax()
print(f"ðŸ•’ HorÃ¡rio com mais abandonos: {hora_abandono}h")
```

```python
# Compras vs. Abandonos
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='comprou', palette='Set2')
plt.title('ComparaÃ§Ã£o entre compras e abandonos')
plt.xlabel('Comprou')
plt.ylabel('Contagem')
plt.show()
```

```python
# Produtos mais abandonados
plt.figure(figsize=(10, 5))
sns.countplot(data=df[df['comprou'] == 'NÃ£o'], y='produto', order=df[df['comprou'] == 'NÃ£o']['produto'].value_counts().index, palette='Reds_r')
plt.title('Produtos mais abandonados')
plt.xlabel('Quantidade de abandonos')
plt.ylabel('Produto')
plt.show()
```

```python
# Compras e abandonos por horÃ¡rio
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='horario', hue='comprou', palette='Set1')
plt.title('Comportamento por hora do dia')
plt.xlabel('Hora do dia')
plt.ylabel('Contagem')
plt.legend(title='Comprou')
plt.show()
```

```python
# INSIGHTS FINAIS
print("\nðŸ“Œ INSIGHTS FINAIS")
print(f"- Taxa de abandono: {taxa_abandono:.2f}%")
print(f"- Produto mais abandonado: {produto_abandonado}")
print(f"- Categoria mais abandonada: {categoria_abandonada}")
print(f"- HorÃ¡rio com mais abandonos: {hora_abandono}h")
print("- ReforÃ§ar aÃ§Ãµes de marketing entre 18h e 21h pode ajudar a recuperar vendas.")
print("- Moda foi a categoria com mais abandonos â€” revisar descriÃ§Ã£o, fotos e preÃ§os pode ser estratÃ©gico.")
```
